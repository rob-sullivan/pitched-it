### Scrape 2048px slide images from webpage

```python
# Missing pitches will have to be searched for manually e.g friend-trusted: https://www.slideshare.net/NoahAtkinson2/friend-trusted-pitch-deck and ring: https://docs.google.com/presentation/d/1PzT_LmToKj0aI8qWnibfjOEs7YHN8nCJ/edit#slide=id.p1

# Load the list of URLs from the CSV file
path = './data/pitches/'

#get all urls from csv file 

with open('./data/startups.csv', 'r', newline='', encoding='utf-8') as csv_file:
    csv_reader = csv.reader(csv_file)
    for row in csv_reader:
        pitch_deck_urls.append(row[0])  # Assuming the URLs are in the first column

#pitch_deck_urls = ['https://www.failory.com/pitch-deck/kong']

for url in pitch_deck_urls:
    response = requests.get(url) #go to website
    main_html = response.text # get site html text

    # Step 2: Parse the main HTML content
    main_soup = BeautifulSoup(main_html, "html.parser") #convert to soup

    #get the name of the link to get name of startup
    folder_name = urlparse(url).path.split('/')[-1]
    os.makedirs(os.path.join(path, folder_name), exist_ok=True)#create a folder if none exist

    # Step 3: Find the iframe element and extract its src attribute
    #check
    iframe = main_soup.find("iframe") #pitch deck may be in an iframe
    iframe_src = iframe["src"] #get links from iframe

    # Step 4: Send a request to the iframe URL
    iframe_response = requests.get(iframe_src) # now go to iframe website
    iframe_html = iframe_response.text #get iframe site html

    # Step 5: Parse the iframe content
    iframe_soup = BeautifulSoup(iframe_html, "html.parser")#convert to soup
    picture_elements = iframe_soup.find_all("picture") # get pictures tag
    if len(picture_elements)>0:

        # Step 6a: Find all <source> elements within <picture> tags
        for picture_element in picture_elements:#get all pictures
            source_elements = picture_element.find_all("source")#get all source tags

            for source_element in source_elements: #within source tag look for srcset
                srcset = source_element.get("srcset")
                if srcset:
                    # Use regex to find URLs that match the pattern
                    pattern = r"https://image.slidesharecdn.com/.+?-2048\.jpg"#find any image that is 2048px rez
                    matches = re.findall(pattern, srcset)
                    
                    for match in matches:
                        url_list = match.split(', ')
                        filtered_matches = [u for u in url_list if "-2048.jpg" in u]

                        img_name = os.path.basename(filtered_matches[0])#get url to download image

                        # Download and save the image in the directory
                        img_response = requests.get(filtered_matches[0])
                        with open(os.path.join(path, folder_name, img_name), "wb") as img_file:
                            img_file.write(img_response.content)

                        print(f"Downloaded image: {img_name}")
    else:
        img_elements = iframe_soup.find_all("img", {"class": "slide-image"})
        for img_element in img_elements:
            srcset = img_element.get("srcset")

            res = ""
            if srcset:
                if '-2048' in srcset:
                    pattern = r"https://image.slidesharecdn.com/.+?-2048\.jpg"#find any image that is 2048px rez
                    res = "-2048.jpg"
                elif '-1024' in srcset:
                    pattern = r"https://image.slidesharecdn.com/.+?-1024\.jpg"#find any image that is 2048px rez
                    res = "-1024.jpg"
                elif '-638' in srcset:
                    pattern = r"https://image.slidesharecdn.com/.+?-638\.jpg"#find any image that is 2048px rez
                    res = "-638.jpg"
                elif '-320' in srcset:
                    pattern = r"https://image.slidesharecdn.com/.+?-320\.jpg"#find any image that is 2048px rez
                    res = "-320.jpg"

                matches = re.findall(pattern, srcset)
                for match in matches:
                    url_list = match.split(', ')
                    filtered_matches = [u for u in url_list if res in u]

                    img_name = os.path.basename(filtered_matches[0])#get url to download image

                    # Download and save the image in the directory
                    img_response = requests.get(filtered_matches[0])
                    with open(os.path.join(path, folder_name, img_name), "wb") as img_file:
                        img_file.write(img_response.content)

                    print(f"Downloaded image: {img_name}")
            else:
                img_src = img_element["src"]
                pattern = r"https://image.slidesharecdn.com/.+?\.jpg"
                # Extract the image file name from the URL
                img_name = os.path.basename(img_src)

                # Extract the image file name from the URL and apply the regex pattern
                img_name = re.search(pattern, img_name).group()

                img_name = img_name.split(".")[0]
                img_name = img_name + ".jpg"

                # Download and save the image in the directory
                img_response = requests.get(img_src)
                with open(os.path.join(path, folder_name, img_name), "wb") as img_file:
                    img_file.write(img_response.content)

    print(f"Images downloaded and saved in '{folder_name}' directory.")
    # Get a list of all folders in the root directory
    folders = [folder for folder in os.listdir(path) if os.path.isdir(os.path.join(path, folder))]

    # Loop through each folder and create a zip file
    for folder in folders:
        folder_path = os.path.join(path, folder)
        zip_filename = os.path.join(path, f"{folder}.zip")  # Specify the full path for the zip file

        # Create a zip file for the folder
        shutil.make_archive(zip_filename[:-4], 'zip', folder_path)

        # Delete the folder and its contents
        shutil.rmtree(folder_path)

        print(f"Zipped and deleted: {folder_path}")
```

```python
# Function to extract data from a startup URL
def extract_data_from_url(url):
    try:
        # Send a GET request to the URL
        response = requests.get(url)
        response.raise_for_status()

        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the parent div with the class 'tools-logo-name-div-block pd'
        pitch_deck_column = soup.find('div', class_='pitch-deck-column-data')

        # Find the card-title div within the parent div
        card_title = pitch_deck_column.find('div', class_='card-title').get_text(strip=True)
        card_description = pitch_deck_column.find('div', class_='card-description').get_text(strip=True)
        pitch_deck_data = {}
        pitch_deck_divs = soup.find_all('div', class_='pitch-deck-data-div')
        #print(pitch_deck_divs)
        for desc_div in pitch_deck_divs:
            pitch_data = desc_div.find_all('div', class_='card-description pitch-deck-data')
            for data in pitch_data:
                if data.find('strong'):
                    strong_tag = data.find('strong').get_text(strip=True)
                else:
                    data_text = data.get_text(strip=True)
            pitch_deck_data[strong_tag] = data_text

            # Extract href from anchor tags if "Links" exists in pitch_deck_data
            if 'Links' in str(desc_div):
                #links_div = soup.find('div', text='Links')
                link_tags = desc_div.find_all('a', href=True)
                links = [link['href'] for link in link_tags]
                pitch_deck_data['Links'] = links

        return {
            'Name': card_title,
            'Description': card_description,
            'Industry': pitch_deck_data.get('Industry', ''),
            'Business Model': pitch_deck_data.get('Business Model', ''),
            'Customer': pitch_deck_data.get('Customer', ''),
            'Round': pitch_deck_data.get('Round', ''),
            'AmountRaised': pitch_deck_data.get('Amount Raised', ''),
            'Investor': pitch_deck_data.get('Investor', ''),
            'Year': pitch_deck_data.get('Year', ''),
            'Links': links if 'Links' in pitch_deck_data else None
        }

    except Exception as e:
        print(f"Error processing URL {url}: {str(e)}")
        return None



# Create or open the CSV file for writing
with open('./data/falory.csv', 'w', newline='', encoding="utf-8") as csv_file:
    fieldnames = ['Name', 'Description', 'Industry', 'Business Model', 'Customer', 'Round', 'AmountRaised', 'Investor', 'Year', 'Links']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    # Write the header row
    writer.writeheader()

    # Read the CSV file containing startup URLs
    with open('./data/startups.csv', 'r', newline='') as url_csv_file:
        reader = csv.reader(url_csv_file)
        next(reader)  # Skip the header row if it exists

        for row in reader:
            startup_url = row[0]
            startup_data = extract_data_from_url(startup_url)
            if startup_data:
                print(startup_data)
                writer.writerow(startup_data)

print("Data written to startup_dataset.csv")
```

### get location from html file

```python
# Define the path to the HTML files and CSV file
html_path = './data/websites/'
csv_file = './data/falory.csv'

# Create a dictionary to store the mapping between Name and Location
name_to_location = {}

# List of existing HTML files in the specified path
html_files = [file for file in os.listdir(html_path) if file.endswith('.html')]

# Loop through HTML files to extract location information
for html_file in html_files:
    #get name from html file name
    name = os.path.splitext(html_file)[0]

    # Load the locally downloaded HTML file
    with open(os.path.join(html_path, html_file), 'r', encoding='utf-8') as file:
        html_content = file.read()

    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the <span> element with the specified attribute
    span_element = soup.find('span', class_='component--field-formatter field-type-identifier-multi', attrs={'_ngcontent-ng-c3823652626': True})

    # Extract text from all <a> tags within the <span> element
    anchor_tags = span_element.find_all('a')

    # Extract the Name and Location (assuming they are in the first and second <a> tags)
    locations = [location.get_text().strip() for location in anchor_tags]
    name_to_location[name] = locations
    print("Location found for " + name)
    """
    if len(anchor_tags) >= 2:
        location = anchor_tags[0].get_text().strip()
        name_to_location[name] = location
        print("Location found for " + name)
    """

# Read the existing CSV file into a DataFrame
df = pd.read_csv(csv_file)

# Add a "Location" column and populate it based on the dictionary
df['Location'] = df['Name'].map(name_to_location)

# Save the updated DataFrame back to the CSV file
df.to_csv(csv_file, index=False)

print("Location data has been added to the CSV file.")
```
