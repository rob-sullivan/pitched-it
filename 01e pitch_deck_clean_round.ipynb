## Startup Rounds

```python
unique_cus = df['Round'].unique()
print(unique_cus)
print("\n")
unique_inv = df['Investor'].unique()
print(unique_inv)
```

```python
round_mapping = {'Pre-Seed': 0, 'Seed': 1, 'Series A': 2, 'Series B': 3, 'Series C': 4, 'Series D': 5, 'Series E': 6, 'Public': 7, 'Crowdfunding': 8, 'Debt Financing': 9}
df['Round'] = df['Round'].replace(round_mapping)
```

```python
category_counts = df['Round'].value_counts()
category_counts = category_counts.sort_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')
plt.title('Startup Round')
plt.xlabel('Round')
plt.ylabel('Number of Startups')
plt.xticks(rotation=0)
plt.show()
```
