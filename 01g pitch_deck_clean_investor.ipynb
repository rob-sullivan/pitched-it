# Pitch Deck Dataset Data Cleaning

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Import Dataset

```python
df = pd.read_csv('data/falory_clean_amount_raised.csv')
```

```python
df.head()
```

```python
unique_inv = df['Investor'].unique()
print(unique_inv)
```

Undisclosed

```python
print(len(df[df['Investor']=='Undisclosed']))
```

```python
df[df['Investor']=='Undisclosed'].head()
```

```python
df = df[df['Investor']!='Undisclosed']
```

```python
unique_inv = df['Investor'].unique()
print(unique_inv)
```

```python
df['Investor'] = df['Investor'].str.replace(r'\s*Investor\s*', '', regex=True)
```

```python
unique_inv = df['Investor'].unique()
print(unique_inv)
```

```python
#VC, Angel Investors, Corporate, Accelerator, Undisclosed, Crowdfunding
#SPAC
investor_dummies  = df['Investor'].str.get_dummies(', ').add_prefix('investor_')
df = pd.concat([df, investor_dummies], axis=1)
```

```python
df.drop('Investor', axis=1, inplace=True)
```

```python
print(df.columns)
```

```python
df.head()
```

```python
new_col_order = [
    'Name',
    'Description',
    'Sector',
    'Business',
    'customer_B2B',
    'customer_B2B2C',
    'customer_B2C',
    'customer_B2G',
    'customer_C2C',
    'Round',
    'investor_Accelerator',
    'investor_Angels',
    'investor_Corporate',
    'investor_Crowdfunding',
    'investor_SPAC',
    'investor_VC',
    'AmountRaisedEuro',
    'Year',
    'Links',
    'Location'
]

# Rearrange the columns
df = df[new_col_order]
```

### save cleaned dataset

```python
# Save the updated DataFrame to a new CSV file or overwrite the original one
df.to_csv('data/falory_clean_investor.csv', index=False)
```
